```python
#Importing the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the data
df = pd.read_csv(r'C:\Users\Yogesh\Desktop\Kaggle_Projects\Euro_2020\euro2020.csv')
```

```python
#Checking the dataset
df.head()
```

```python
#Shape of the data - rows and columns
df.shape
```

```python
df.columns
```

```python
#To check the mean, median and other percentiles of the numerical data
df.describe()
```

```python
#Checking the number of Null values in each column
df.isnull().sum()
```

```python
#Checking the percentage of Null values in each column
df.isnull().sum()/len(df)*100
```

```python
#Changing the Position column's acronyms
df['Position']=df['Position'].map({'F':'Forward','M':'Midfielder','D':'Defender'})
```

```python
#Imputing the missing value of Position column with Mode
df['Position']=df['Position'].fillna(df['Position'].mode()[0])
```

```python
#Replacing the rest of the missing values with Zeros, as it is possible that the rest of the values are zero.
#For example : Right foot goals of a player can be zero, Assists can be zero etc.
df = df.replace(np.NaN, 0)
```

```python
#Cross checking if all the abovementioned steps are completed 
df.head(5)
```

```python
#Cross cehcking the percentage of Null values - Now they are all 0
df.isnull().sum()/len(df)*100
```

### Data Visualization

#### > Univariate Analysis:

```python
#Pie chart for the percentage of Positions in the dataset

plt.figure(figsize=(9,7))
df.Position.value_counts().plot(kind='pie', autopct='%.2f',colors=['tomato', 'gold', 'skyblue'])
plt.show()
```

##### Most of the player in the dataset play in the Position of Midfielders

```python
#Pie chart for the percentage of Average Goals per game in the dataset

plt.figure(figsize=(12,10))
df['Avg gpg'].value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
```

##### In most of the games the average goal scored was 0

```python
#Comparing the various percentiles of  'On target' and 'Off target'

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(x='On target',data=df)
plt.subplot(1,2,2)
sns.boxplot(x='Off target',data=df)
plt.show()
```

##### Seems like the On and Off targets have similar median, but at 75% the On target seems better.

#### > Bivariate Analysis

```python
# Players vs Goals
#Where goals are greater than Zero - 0
df1 = df[df['Goals']>0]

#Plotting a barplot
plt.figure(figsize=(20,15))
plt.title('Total goals by players', fontsize=20)
sns.barplot(x='Player',y='Goals',data=df1, order=df1.sort_values('Goals',ascending = False).Player)
plt.xticks(rotation = 90)
plt.show()
```

##### Again Ronaldo is a top scorer with most number of goals

```python
# Country vs Goals
#Where goals are greater than Zero - 0

#Plotting a barplot
plt.figure(figsize=(15,8))
plt.title('Total goals by Countries', fontsize=20)
sns.barplot(x='Country',y='Goals',data=df1 ,estimator=sum, ci=None)
plt.xticks(rotation = 90)
plt.show()
```

##### England owns the first place in most number of goals

```python
# Right foot goals vs Left foot goals
plt.figure(figsize=(14,6))
sns.set_theme(style="whitegrid")
plt.subplot(1,2,1)
plt.title("Right foot goals", fontsize=20)
sns.barplot(x='Right foot goals',y='Goals',data=df ,estimator=sum, ci=None)
plt.subplot(1,2,2)
plt.title("Left foot goals", fontsize=20)
sns.barplot(x='Left foot goals',y='Goals',data=df ,estimator=sum, ci=None)
plt.show()
```

##### Right foot seems to perform better.

```python
# Player vs Assists
#Where Assists are greater than Zero - 0
df_Assists = df[df['Assists']>0]

plt.figure(figsize=(10,8))
sns.set_theme(style="whitegrid")
plt.title('Total Assists by players', fontsize=20)
sns.barplot(x='Player',y='Assists',data=df_Assists ,estimator=sum, ci=None,order=df_Assists.sort_values(by=['Assists'], ascending=False).Player)
plt.xticks(rotation = 90)
plt.show()
```

##### Seteve Zuber takes the crown with most assists

```python
# Country vs Yellow cards

plt.figure(figsize=(10,6))
sns.set_theme(style="whitegrid")
sns.barplot(x='Country',y='Yellow cards',data=df,estimator=sum, ci=None)
plt.xticks(rotation = 90)
plt.show()
```
